```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Defining the function to generate the points required for the graph
def bifurcate(P, X, Y):
  for a in P:
    m = np.random.random()
    for n in range(1051):
        m = (a * m) * (1 - m)
        # removing transient values
        if n > 500:
            X.append(a)
            Y.append(m)
```

# General Bifurcation diagram

```python
P = np.linspace(0, 4.0001, 2000)

X = []
Y = []

bifurcate(P, X, Y)

plt.figure(figsize=(10, 6))
plt.plot(X, Y, ls='', marker=',', markersize=1)
plt.xlabel('Parameter (a)')
plt.ylabel('Final Values')
plt.title('Bifurcation Diagram of Logistic Map')
plt.show()
```

## Zoomed version of bifurcation diagram from a ∈  [0,1)

```python
P = np.linspace(0, 1, 2000)

X = []
Y = []

bifurcate(P, X, Y)

plt.figure(figsize=(8, 4))
plt.plot(X, Y, ls='', marker=',', markersize=1)
plt.xlabel('Parameter (a)')
plt.ylabel('Final Values')
plt.title('Zoomed Bifurcation Diagram for 0 <= a < 1')
plt.show()
```

### Why do all the values remain at 0 all the way till 1?
For 0 ≤ a < 1, the logistic map function f(x)=a⋅x⋅(1−x) tends to drive the values of x towards zero for most initial conditions. This is a ***stable fixed point***, and initial values converge to this point, leading to the concentration of values at zero in the bifurcation diagram. The mathematical explanation lies in the stability properties of the logistic map for these parameter values.

## Zoomed version of bifurcation diagram from a ∈  [1,3)

```python
P = np.linspace(1, 3, 2000)

X = []
Y = []

bifurcate(P, X, Y)

plt.figure(figsize=(8, 4))
plt.plot(X, Y, ls='', marker=',', markersize=1)
plt.xlabel('Parameter (a)')
plt.ylabel('Final Values')
plt.title('Zoomed Bifurcation Diagram for 1 <= a < 3')
plt.show()
```

### Why are the values not concentrated at 0 now?
Once the a value crosses 1, out of the two fixed points -> 0 and (a-1)/a, the ***stable*** fixed point is now (a-1)/a. The trajectory eventually converges to around 0.66 = 2/3. As we can see, this is following the equation mentioned (a-1)/a.

## Zoomed version of Bifurcation Diagram from a ∈  [3 , 3.4]

```python
P = np.linspace(3, 3.4, 2000)

X = []
Y = []

bifurcate(P, X, Y)

plt.figure(figsize=(8, 4))
plt.plot(X, Y, ls='', marker=',', markersize=1)
plt.xlabel('Parameter (a)')
plt.ylabel('Final Values')
plt.title('Zoomed Bifurcation Diagram for 3 <= a <= 3.4')
plt.show()
```

- The bifurcation diagram for 3 ≤ a ≤ 3.4 will reveal the dynamic behavior of the logistic map within this parameter range. You can observe the emergence of periodic windows and the onset of chaos.
- As a increases within this range, the logistic map undergoes period-doubling bifurcations, leading to the creation of stable fixed points and periodic orbits.
- Period-two orbits occur when the trajectory alternates between two values. These orbits correspond to the doubling of the period of the previous stable orbit.
- To determine initial conditions leading to period-two orbits, you can observe the locations in the bifurcation diagram where the system transitions from a stable orbit to a period-two orbit. As we can see a=3 is the point at which the period doubling occurs, and the next one will occur at a=3.4.
- Stability of period-two fixed points can be assessed similarly to fixed points. It involves analyzing the behavior of the derivative at these points. Stable period-two orbits have a derivative with an absolute value less than 1.

## Trajectory for a = 3.5 and x0 = 0.012 for 40 iterations.

```python
def logistic_map(a, x, iterations):
    trajectory = [x]
    for _ in range(iterations - 1):
        x = a * x * (1 - x)
        trajectory.append(x)
    return trajectory

a = 3.5
x0 = 0.012
iterations = 40

trajectory = logistic_map(a, x0, iterations)

plt.plot(range(1, iterations + 1), trajectory, marker='o')
plt.xlabel('Iteration')
plt.ylabel('Value')
plt.title('Logistic Map Trajectory for a = 3.5, x0 = 0.012')
plt.show()
```

### Observations
- The points seem to stabilise into a period 4 orbit, which matches our observations from previous plot
- The individual points seem to be getting further and further spaced out however, as the number of iterations increase

## Trajectory for a = 3.5 and x0 = 0.012 for 40 iterations.

```python
a = 1 + 2 * np.sqrt(2)
x0 = 0.012
iterations = 40

trajectory = logistic_map(a, x0, iterations)

plt.plot(range(1, iterations + 1), trajectory, marker='o')
plt.xlabel('Iteration')
plt.ylabel('Value')
plt.title(f'Logistic Map Trajectory for a = {a:.3f}, x0 = {x0}')
plt.show()
```

### Observations
- The points seem to be completely random, no pattern immediately stands out
- This would imply that at this value of a, the system has become chaotic.

## Why are we limiting our a values to be less than 4?
If a exceeds 4, the logistic map loses its stability, and the dynamics become increasingly unpredictable. The system can enter regions of parameter space where the trajectory does not converge, making it challenging to analyze and understand the behavior.
