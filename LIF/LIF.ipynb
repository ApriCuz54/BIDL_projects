# Leaky Integrate and Fire Model

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
R = 10
tau = 10
V0 = -70
Vth = -55
V_spike = 20
V_reset = -73
step_size = 0.1
```

```python
t = np.arange(0, 100, step_size)
I = np.zeros(len(t))
for i in range(100, 501):
  I[i] = 1
Vm = np.zeros(len(t))
Vm[0] = V0
```

```python
iter =1
while iter < len(t):
  if Vm[iter-1] >= Vth:
    Vm[iter] = V_spike
    Vm[iter+1] = V_reset
    iter = iter+2
  else:
    Vm[iter] = Vm[iter-1] + step_size*(((V0-Vm[iter-1])/tau) + I[iter-1]*R/tau )
    iter = iter+1
```

```python
plt.plot(t, Vm)
plt.plot(t, Vth*np.ones(len(t)),'--')
plt.xlabel("time (ms)")
plt.ylabel("Membrane Potential (mV)")
```

## Observations
- The membrane potential was not able to reach the threshold level, and hence no spike was generated.

- As the current given is not of enough magnitude, the spikes seems to be plateauing as it gets closer to 50ms.

- Once the current is stopped, the membrane potential sharply decreases and levels out at the resting potential value.

- This can be considered to be a failed initiation of a neuron.
